```python
in_root = '/orcd/data/lhtsai/001/om2/mabdel03/files/ACE_Analysis/Analysis/Tsai/Processing/ACE/Final_Pipeline/QC/Doublets'

out_root = '/orcd/data/lhtsai/001/om2/mabdel03/files/ACE_Analysis/Analysis/Tsai/Processing/ACE/Final_Pipeline/Batch_Correction'

projids = os.listdir('/orcd/data/lhtsai/001/om2/mabdel03/files/ACE_Analysis/Analysis/Tsai/Processing/ACE/Final_Pipeline/QC/Doublets')
```

```python
len(projids)
```

```python
obj_list = []
projid_list = []
counter = 0


for projid in projids:
    try:
        file_path = os.path.join(in_root, projid, projid + '_doublet_filtered.h5')
        obj = sc.read(file_path)
        obj_list.append(obj)
        projid_list.append(projid)
        counter += 1
        if counter == 20:
            break
    except FileNotFoundError:
        print(f'{projid} does not exist')
for obj, projid in zip(obj_list, projid_list):

	obj.obs['batch'] = [projid for i in range(len(obj.obs))] 

adata = ad.concat(obj_list)
```

```python
scvi.model.SCVI.setup_anndata(
    adata,
    layer="counts",
    batch_key="batch"
)
```

```python
vae = scvi.model.SCVI(adata)
```

```python
vae.train()
```

```python
adata.obsm["X_scVI"] = vae.get_latent_representation()
```

```python
adata = sc.read('/net/vast-storage/scratch/vast/lhtsai/mabdel03/files/ACE_Analysis/Analysis/Tsai/Processing/ACE/Final_Pipeline/Batch_Correction/batch_corrected.h5ad')

import scipy.sparse as sp
import anndata as ad
import numpy as np

# Convert the main data matrix (X) to a sparse matrix
if not sp.issparse(adata.X):
    adata.X = sp.csr_matrix(adata.X)

# Convert all layers to sparse matrices
for layer in adata.layers:
    if not sp.issparse(adata.layers[layer]):
        adata.layers[layer] = sp.csr_matrix(adata.layers[layer])

# Convert all obsm matrices to sparse
for key in adata.obsm:
    if isinstance(adata.obsm[key], np.ndarray):
        adata.obsm[key] = sp.csr_matrix(adata.obsm[key])

# Convert all varm matrices to sparse
for key in adata.varm:
    if isinstance(adata.varm[key], np.ndarray):
        adata.varm[key] = sp.csr_matrix(adata.varm[key])
adata.write('corrected_data_sparse.h5ad')
```
